**Linear Regression using Single Varible in Machine Learning**

Predicting the Premium of person based on Age using available data with help Pandas, SkLearn and Seaborn Libraries.

Learn from given YT video : https://www.youtube.com/watch?v=PcUnYSxAtAo



```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv(r'D:/Coding/Learn_Python/Data/Insurance_Premeuim_Single_Var.csv')
```

```python
df
```

```python
sns.lmplot(x="age", y="premium", data=df)
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["age"]],df[["premium"]])
```

At this moment, our model has been train based on given data. Now we can use predict function to  find out the unknown value of Premium for any Age we want. 

```python
reg.predict([[21]])
```

As we can see, at Age 21 a might need to pay yearly premium of Rs. 13480

```python
reg.predict([[50]])
```

As we can see, at Age 50 a might need to pay yearly premium of Rs. 65100

For a line Y = mX + c, Where m is Slope and c is Intercept

```python
reg.coef_
```

```python
reg.intercept_
```
