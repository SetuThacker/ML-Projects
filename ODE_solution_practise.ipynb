```python
import numpy as np
from scipy.integrate import odeint
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
func = lambda y,x : x + y/5

y0 = -3

x1 = np.linspace(0,4,101)

y = odeint(func, y0, x)

plt.plot(x,y,'--', color = "red", lw = 1, marker="+", markerfacecolor = "black")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
def dvdt(t,v):
    return 3*(v**2) - 5 # dvdt = 3(v^2) - 5 which we got from by just simplifying the differntial equations.
v0 = 0 # Write down your intial condition.

```

```python
t = np.linspace(0,1,100)
sol_m1 = odeint(dvdt, y0 = v0, t = t, tfirst= True)
sol_m2 = solve_ivp(dvdt, t_span = (0, max(t)), y0 = [v0], t_eval = t)
```

```python
sol_m1.T[0]
```

If we have more variable than t, then we need to use other number than 0.

```python
sol_m2.y
```

```python
v_sol_m1 = sol_m1.T[0]
v_sol_m2 = sol_m2.y[0]
```

```python
plt.plot(t,v_sol_m1, label = "Solution using odeint")
plt.plot(t,v_sol_m2,"--", label = "Solution using solv_ivp")
plt.ylabel("$v(t)$", fontsize = 22)
plt.xlabel("$t$", fontsize = 22)
plt.legend()
plt.show()
```
